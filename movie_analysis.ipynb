```python
from pandas import Series,DataFrame 
```

```python
split=movies_genre['director'].str.split('|').apply(Series,1).stack()
# assign the row to the right place 
split.index= split.index.droplevel(-1)
split.name='director_split'
del movies_genre['director']
movies_genre =movies_genre.join(split)
```

```python
movies_genre
```

```python
split=movies_genre['genres'].str.split('|').apply(Series,1).stack()
# assign the row to the right place 
split.index= split.index.droplevel(-1)
split.name='genres_split'
del movies_genre['genres']
movies_genre =movies_genre.join(split)
```

Check which type of genre is popular in each season ?

```python
movies_genre
```

```python
movies_genre['release_date'] = pd.to_datetime(movies_genre['release_date'])
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
```

```python
movies_genre['season'] = movies_genre['release_date'].dt.month.apply(get_season)
```

```python
season_genre_revenue = movies_genre.groupby(['season', 'genres_split'])['revenue'].mean().reset_index()
```

```python
most_popular_genre_per_season = season_genre_revenue.loc[season_genre_revenue.groupby('season')['revenue'].idxmax()]

most_popular_genre_per_season
```
